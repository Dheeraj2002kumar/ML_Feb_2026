```python
import csv

fn = "placement.csv"
fields = []
rows = []

with open(fn, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    fields = next(csvreader)

    for row in csvreader:
        rows.append(row)
    print("Total no. of rows: %d", (csvreader.line_num))

print('Field names are: ', ', '.join(field for field in fields))

# printing first 5 rows
print('\nFirst 5 rows are:\n')

for row in rows[:5]:
    #parsing each column of a row
    for col in row:
        print("%10s" % col, end=" "),
    print("\n")
```

```python
import matplotlib.pyplot as plt

names = ['A', 'B', 'C']
values = [19, 50, 29]
values_2 = [27, 15, 34]

fig = plt.figure(figsize=(8.0, 6.0))

plt.text(0, 30, r'Plot text like this', fontsize=15, horizontalalignment='center')

plt.xlabel("Label for X")
plt.ylabel("Label for Y")
plt.bar(names, values)
plt.legend(['Data'], loc="upper right")
plt.suptitle('Test Plots')
plt.show()
```

```python
# Fruits Yeild
years = range(2000, 2006)
apples = [100, 123, 98, 112, 156, 130]
oranges = [102, 130, 78, 90, 140, 129]

plt.xlabel('Years')
plt.ylabel('Yield in Tons')
plt.title('Crop Yeild in Maharashtra')
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples)
plt.show()
```
